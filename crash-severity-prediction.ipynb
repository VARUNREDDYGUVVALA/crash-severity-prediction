```python
import pandas as pd

# Path to the dataset
file_path = "D:\Drive\data mining\Traffic_Crashes_-_Crashes (1).csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Display the shape and first few rows
print("Shape of the dataset:", df.shape)
df.head()
```

```python
# Columns to drop
cols_to_drop = ['CRASH_RECORD_ID', 'CRASH_DATE_EST_I', 'CRASH_DATE', 'LATITUDE', 'LONGITUDE', 'LOCATION']

# Drop the columns
df.drop(columns=cols_to_drop, inplace=True)

# Confirm the change
print("Remaining columns:", df.columns.tolist())
```

```python
df.drop(['DATE_POLICE_NOTIFIED','PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I'], axis=1, inplace=True)
```

```python
df.drop(['LANE_CNT'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.drop(['STREET_NO'],axis=1,inplace=True)
```

```python
df['CRASH_TYPE'].value_counts()
```

```python
# Drop rows with null 'CRASH_TYPE'
df = df[df['CRASH_TYPE'].notna()]

# Sample 20,000 rows from each class
sampled_df = df[df['CRASH_TYPE'] == 'NO INJURY / DRIVE AWAY'].sample(n=20000, random_state=42)
sampled_df = pd.concat([
    sampled_df,
    df[df['CRASH_TYPE'] == 'INJURY AND / OR TOW DUE TO CRASH'].sample(n=20000, random_state=42)
])

# Shuffle the resulting dataframe
sampled_df = sampled_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Verify the result
print(sampled_df['CRASH_TYPE'].value_counts())
```

```python
sampled_df.isnull().sum()
```

```python
# Set threshold for dropping columns
threshold = 5000

# Drop columns with null values exceeding the threshold
cols_to_drop_nulls = sampled_df.columns[sampled_df.isnull().sum() > threshold]
df_cleaned = sampled_df.drop(columns=cols_to_drop_nulls)

# Show dropped columns and shape after cleaning
print("Dropped columns due to high null values:", list(cols_to_drop_nulls))
print("New shape of DataFrame:", df_cleaned.shape)
```

```python
df_cleaned.info()
```

```python
# Separate numeric and categorical columns
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns

# Impute missing values
for col in numeric_cols:
    if df_cleaned[col].isnull().sum() > 0:
        median_value = df_cleaned[col].median()
        df_cleaned[col].fillna(median_value, inplace=True)

for col in categorical_cols:
    if df_cleaned[col].isnull().sum() > 0:
        mode_value = df_cleaned[col].mode()[0]
        df_cleaned[col].fillna(mode_value, inplace=True)

# Confirm no nulls remain
print("Any null values left?", df_cleaned.isnull().any().any())
```

```python
from scipy.stats import pointbiserialr
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Ensure that the 'CRASH_TYPE' column is converted to a numeric dtype if it's categorical
if df['CRASH_TYPE'].dtype == 'object':
    df['CRASH_TYPE'] = LabelEncoder().fit_transform(df['CRASH_TYPE'])

# Correlation of continuous variables with CRASH_TYPE (using Point-Biserial Correlation)
correlations = {}
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    if col != 'CRASH_TYPE':  # Exclude target variable itself
        # Handle potential non-numeric values gracefully
        if not np.issubdtype(df[col].dtype, np.number):
            continue  # Skip non-numeric columns
        corr, _ = pointbiserialr(df[col], df['CRASH_TYPE'])
        correlations[col] = corr

# Sort the correlations by absolute value, highest first
sorted_correlations = sorted(correlations.items(), key=lambda x: abs(x[1]), reverse=True)

# Plot the top 10 most correlated variables
top_n = 10
top_correlations = sorted_correlations[:top_n]

# Prepare data for plotting
top_columns, top_corr_values = zip(*top_correlations)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=list(top_columns), y=list(top_corr_values), palette='viridis')
plt.title(f'Top {top_n} Variables Correlated with CRASH_TYPE')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Model Building and evaluation

```python
# Label encode categorical features
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Split features and target
X = df.drop('CRASH_TYPE', axis=1)
y = df['CRASH_TYPE']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(100,), max_iter=300)
}

# Evaluate each model
results = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1]

    # Confusion matrix
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f"Confusion Matrix - {name}")
    plt.tight_layout()
    plt.show()

    # ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {name}')
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    # Classification report
    print(f"--- {name} ---")
    print(classification_report(y_test, y_pred))
    results[name] = roc_auc
```

```python
# Assuming 'results' contains model names and their AUC scores
for model_name, auc_score in results.items():
    print(f"AUC Score for {model_name}: {auc_score:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter grids
param_grids = {
    "Logistic Regression": {
        "C": [0.01, 0.1, 1, 10],
        "solver": ['liblinear']
    },
    "Random Forest": {
        "n_estimators": [100, 200],
        "max_depth": [None, 10, 20]
    },
    "Gradient Boosting": {
        "n_estimators": [100, 200],
        "learning_rate": [0.05, 0.1],
        "max_depth": [3, 5]
    },
    "AdaBoost": {
        "n_estimators": [50, 100],
        "learning_rate": [0.5, 1.0]
    },
    "Neural Network": {
        "hidden_layer_sizes": [(50,), (100,), (50, 50)],
        "alpha": [0.0001, 0.001]
    }
}

# Initialize models again for tuning
base_models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Neural Network": MLPClassifier(max_iter=300)
}

tuned_results = {}
best_params = {}

# Run GridSearchCV for each model
for name, model in base_models.items():
    grid = GridSearchCV(model, param_grids[name], cv=5, scoring='roc_auc', n_jobs=-1)
    grid.fit(X_train_scaled, y_train)

    # Predict and evaluate
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test_scaled)
    y_proba = best_model.predict_proba(X_test_scaled)[:, 1]

    # ROC AUC
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)

    tuned_results[name] = roc_auc
    best_params[name] = grid.best_params_

    # Print evaluation
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f"Tuned Confusion Matrix - {name}")
    plt.tight_layout()
    plt.show()

    plt.figure()
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Tuned ROC Curve - {name}')
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    print(f"--- {name} (Tuned) ---")
    print("Best Parameters:", grid.best_params_)
    print(classification_report(y_test, y_pred))
```
